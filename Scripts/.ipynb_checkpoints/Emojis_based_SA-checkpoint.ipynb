```python
def emotidetect(text): 
    er = adv.extract_emoji(text)
    d=er['emoji']
    #remove empty list in list
    s = [x for x in d if x]
    #Select first item in list
    if s:
        ft =s[0]
        if ft in PosSent:
            Sent ='POS'
        elif ft in NegSent:
            Sent ='NEG'
        elif ft in NeutSent:
            Sent = 'NEU'
        elif ft:
            Sent='NEG'
    else :
        Sent ='UNK'
    return Sent
```
